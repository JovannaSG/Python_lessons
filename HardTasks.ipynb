### Все вхождения подстроки в строку 
Напишите программу, которая находит все вхождения подстроки в строку. Строка и подстрока вводятся с клавиатуры.

<details>
<summary>Ответ</summary>
<code>
s: str = input("Введите строку: ")
substring: str = input("Введите подстроку: ")

positions: list = []
pos = s.find(substring)
while pos != -1:
    positions.append(pos)
    pos = s.find(substring, pos + 1)

print(positions)
</code>
</details>

### Найти два максимума в массиве

Напишите программу, которая находит и выводит на экран два максимума в массиве целых чисел. Без сортировки массива.

<details>
<summary>Ответ</summary>
<code>
numbers: list[int] = [12, 2, 5, 7, 1]
max1: int = numbers[0]
max2: int = numbers[1]
for number in numbers:
    if number > max1:
        max2 = max1
        max1 = number
    elif number > max2 and number < max1:
        max2 = number
print(max1, max2)
</code>
</details>

### Задача Иосифа Флавия

Формат входных данных
На вход программе подаются два натуральных числа n и k, каждое на отдельной строке.

Формат выходных данных
Программа должна вывести одно число – номер человека, который останется в кругу последним.

### Зодиак

Китайский гороскоп назначает животным годы в 12-летнем цикле. Один 12-летний цикл показан в таблице ниже. Таким образом, 2012 год будет очередным годом дракона.

|Год|Животное|
|---|--------|
|2000|Дракон|
|2001|Змея|
|2002|Лошадь|
|2003|Овца|
|2004|Обезьяна|
|2005|Петух|
|2006|Собака|
|2007|Свинья|
|2008|Крыса|
|2009|Бык|
|2010|Тигр|
|2011|Заяц|

Напишите программу, которая считывает год и отображает название связанного с ним животного. Ваша программа должна корректно работать с любым годом, не только теми, что перечислены в таблице.

Формат входных данных
На вход программе подается натуральное число – год.

Формат выходных данных
Программа должна вывести текст – название животного.

### Standard American Convention

На вход программе подается натуральное число. Напишите программу, которая вставляет в заданное число запятые в соответствии со стандартным американским соглашением о запятых в больших числах.

Формат входных данных
На вход программе подается натуральное число $n$ $(0 < n < 10^{100})$.

Формат выходных данных
Программа должна вывести число с запятыми в соответствии с условием задачи.

|Входные данные|Выходные данные|
|--------------|---------------|
|1000000|1,000,000|
|100|100|
|12345|12,345|

<details>
<summary>Ответ</summary>
<code>
n = int(input())

print(f"{n:,}")
</code>
</details>
