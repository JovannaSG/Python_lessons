### Переменные

Переменные предназначены для хранения данных.

ОЗУ можно представить как шкаф, в котором очень много выдвигаемых полочек, каждая полочка подписана (имеет имя) и хранит какое-то значение.

````python
name: str = "Tom"
Name: str = "John"

print(name)
print(Name)
````

#### Типы данных

* bool - логический тип данных (True, False)
* int - целые числа (2, 4, -223, 10987)
* float - вещественные числа (1.23, -555.124215)
* str - строка ("abc", 'Hello', "44")



#### Тип int

В python есть динамическая типизация, поэтому писать тип переменной не обязательно

Арифметические операции с числами:

- \+ - сложение
- \- - вычитание
- \* - умножение
- \/ - деление
- \// - целочисленное деление
- \% - получение остатка от деления
- \** - возведение в степень

````python
a: int = 6
b = 4  # b: int

print(a * b)
print(a - b)
print(a + b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
````

````python
a: float = 6.0
b = 4.2

print(a * b)
print(a - b)
print(a + b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)
````

#### Булевы значения как числа

Логические значения в Python можно трактовать как числа. Значению True соответствует число 1, в то время как значению False соответствует 0. Таким образом, мы можем сравнить логические значения с числами:

Приведенный ниже код:

```python
print(True == 1)
print(False == 0)
```

выводит:

```bash
True
True
```

Мы можем также применять арифметические операции к логическим значениям.

Приведенный ниже код:

```python
print(True + True + True - False)
print(True + (False / True))
```

выводит:

```bash
3
1.0
```

Возможность трактовать булевы выражения как числа на практике используется не так часто. Однако есть один прием, который может оказаться полезным. Поскольку True равно 1, а False равно 0, сложение логических значений вместе – это быстрый способ подсчета количества значений True. Это может пригодиться, когда требуется подсчитать количество элементов, удовлетворяющих условию.

```python
numbers = [1, 2, 3, 4, 5, 8, 10, 12, 15, 17]
res = 0

for num in numbers:
    res += (num % 2 == 0)

print(res)
```

выводит количество четных элементов списка numbers, то есть число 5.

### Как узнать тип данных, хранящихся в переменных?

Для того, чтобы узнать тип переменной используется метод (команда) `type()`

````python
a = 4
b = 3.4
c = "avc"

print(type(a))
print(type(b))
print(type(c))
````

### Консольный ввод и вывод

Вывод в консоль.

Мы уже ранее использовади с вами метод (команду) `print()`, которая выводила на экран содержимое переменных и результаты арифметических операций. Важно понимать, что `print()` все преобразует в строки (str).

Консольный ввод (ввод с клавиатуры)

Для ввода значений с клавиатуры используется метод (команда) `input()`. Здесь тоже есть нюанс, все что вы введете в консоль, становится `str` (строкой).

````python
print("Введите любое значение:")
input_value: str = input()

a: int = int(input("Введите число а (int): "))
b: float = float(input("Введите число b (float): "))

print(input_value, type(input_value))
print(a, type(a))
print(b, type(b))
````
